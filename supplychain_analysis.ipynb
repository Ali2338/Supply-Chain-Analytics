```python
import pandas as pd

df = pd.read_excel("Retail-Supply-Chain-Sales-Dataset.xlsx")
print(df.shape)
df.head(10)
```

```python
df.columns = df.columns.str.lower().str.strip().str.replace(" ","_")
df = df.rename(columns={"sub-category":"sub_category"})
df.columns
```

```python
df['order_date'] = pd.to_datetime(df['order_date'])
df['ship_date'] = pd.to_datetime(df['ship_date'])
```

```python
df['delivery_delay_days'] = (df['ship_date'] - df['order_date']).dt.days
df[['order_date','ship_date','delivery_delay_days']].head(10)
```

```python
df['late_delivery'] = df['delivery_delay_days'].apply(lambda x : 'Yes' if x > 7 else 'No')
df[['delivery_delay_days','late_delivery']].head(20)
```

```python
df = df.rename(columns={'retail_sales_people':'supplier'})
```

```python
supplier_perf = df.groupby('supplier').agg({
    'delivery_delay_days':'mean',
    'order_id':'count',
    'sales':'sum'
}).reset_index()
supplier_perf.columns = ['supplier','avg_delivery_delay','total_orders','total_sales']
supplier_perf.head(5)
```

```python
warehouse_load = df.groupby('region').agg({
    'quantity':'sum',
    'sales':'sum'
}).reset_index()

warehouse_load.columns = ['warehouse_region', 'total_quantity', 'total_sales']

warehouse_load.head()
```

```python
df.to_csv("order_cleaned.csv")
supplier_perf.to_csv("supplier_performance.csv")
warehouse_load.to_csv("warehouse_load.csv")
```
